```python
t1=df['MinTemp']
t2=df['MaxTemp']
plt.figure(1)
plt.subplot(211)
plt.plot(t1,t1,'bo',label='t1')
plt.plot(t2,t2**2,'r',label='t2')
plt.legend()

plt.subplot(212)
plt.subplot(t2,np.cos(2*np.pi*t2),'r--')
plt.grid(True)
plt.text(2,18,0.8,r'cos func')
plt.show()
```
