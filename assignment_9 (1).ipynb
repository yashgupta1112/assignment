```python
#QUESTION 1
#Importing the required libraries 
import re
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
#Importing datasets
df=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/train.csv")
df
```

```python
##Importing datasets
df1=pd.read_csv("https://raw.githubusercontent.com/Shreyas3108/Titanic-EDA-and-Survival-prediction/master/test.csv")
df1
```

```python
##Concatenate
frames=[df,df1]
result=pd.concat(frames)
result
```

```python
result.head()
```

```python
##Remove NaN values from the column ‘Survived’. 
ab=result.dropna(subset=["Survived"])
ab
```

```python
ab.isnull().sum()
```

```python
##Create a new column 'Has_Cabin' that tells you whether passengers had a cabin or not. 
ab.insert(loc=2,column="Has_Cabin",value=ab["Cabin"].notna())
ab
```

```python

##Create a new column “Title”, which will contain the titles of the passengers such as Mr., Mrs., etc. 
##There should be five titles: Mr, Mrs, Master, Miss, and Special. 


ab["title"]=ab.Name.apply(lambda x:re.search('([A-Z][a-z]+)\.',x).group(1))
ab["title"].unique()
```

```python
ab
```

```python
ab["title"]=ab["title"].replace(["Don","Rev","Dr","Major","Lady","Sir","Col","Capt","Countess","Jonkheer"],"special")
ab["title"]
```

```python
##Create two new columns “Categorical_age” and “Categorical_fare”,

ab["Categorical_age"]=pd.qcut(ab.Age,q=4,labels=False)
ab["Categorical_fare"]=pd.qcut(ab.Fare,q=4,labels=False)
ab.head()
```

```python
##Impute the missing values using .fillna() method.

ab=ab.fillna(ab.median())
ab
```

```python
ab.isnull().sum()
```

```python
###Drop columns Cabin, Name, Passengerid, Ticket, Age, Fare, SibSp, Parch.

sc=ab.drop(["Cabin","Name","PassengerId","Ticket","Age","Fare","SibSp","Parch"],axis=1)
sc
```

```python
##Transform all variables into numeric ones using get_dummies().

sc = pd.get_dummies(sc)
```

```python
sc
```

```python

X = sc.drop(["Survived"],axis = 1)
Y=sc["Survived"]
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.3)
```

```python
reg = LogisticRegression()
reg.fit(X_train,Y_train)
```

```python
pred = reg.predict(X_test)
```

```python
result = pd.DataFrame({"Predicted":pred,"Actual":Y_test})
result
```

```python
reg.score(X_test,Y_test)
```
